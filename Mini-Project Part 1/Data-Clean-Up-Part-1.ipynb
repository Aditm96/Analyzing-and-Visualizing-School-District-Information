```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
File = pd.read_csv(file, low_memory = False)
File_df = pd.DataFrame(File)
File_df
```

```python
# Inspect all columns
File_df.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Use iloc to accomplish this (remember that you need to take the position of the rows and columns into account)
File_df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
# Hint use the df.replace function
File_df.replace(to_replace=[0.0, 1.0], value=["No", "Yes"], inplace = True)
File_df
```

```python
# Calculate total number of respondents in survey
# consider using the len() function
Total = len(File_df["Age"])
Total
```

```python
# Extract rows corresponding only to people who attended a bootcamp
# Filter using df.loc on the AttendedBootcamp column
Attended = File_df.loc[File_df["AttendedBootcamp"] == "Yes", :]
Attended
```

```python
# Calculate average age of attendees
# Pull out the `Age` series and take the average of it
Average_Age = round(Attended.Age.mean(), 2)
Average_Age
```

```python
# Calculate how many people attended a bootcamp
# Count the values of the AttendedBootcamp column
Population = len(Attended["Age"])
Population
```

```python
# Calculate how many attendees hold degrees
# It is easier to do this with `value_counts() count values for the SchoolDegree column
Degree = Attended.SchoolDegree.value_counts()
Degree
```

```python
# Count the number of records where the person is a degree holder
# There are several ways to approach this. You can look for people who have degrees
# or for people who don't have degrees depending on the value of the SchoolDegree column
degree_Holder = Attended.loc[
                            (Attended['SchoolDegree'] == "bachelor's degree")|
                            (Attended['SchoolDegree'] == "master's degree (non-professional)")|
                            (Attended['SchoolDegree'] == "professional degree (MBA, MD, JD, etc.)")|
                            (Attended['SchoolDegree'] == "Ph.D.")|
                            (Attended['SchoolDegree'] == "associate's degree"), :]
#Degree_Holder
degree_holder = degree_Holder['SchoolDegree'].count()
degree_holder
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
Male_Count = len(Attended.loc[(Attended['Gender'] == 'male'), :])
Female_Count = len(Attended.loc[(Attended['Gender'] == 'female'), :])
Non_Specific = len(Attended.loc[(Attended['Gender'] == 'genderqueer')|(Attended['Gender'] == 'trans')|
                            (Attended['Gender'] == 'agender'), :])
Sum = Male_Count + Female_Count + Non_Specific
print(f'Total: {Sum}')
print(f'Male: {Male_Count}')
print(f'Female: {Female_Count}')
print(f'non_specific: {Non_Specific}')
```

```python
# Calculate percentage of respondents who attended a bootcamp
Attend = len(Attended["Age"])
Percentage = (Attend / Total)*100
Percentage
```

```python
# Calculate percentage of respondents belonging to each gender
Male_Percentage = (Male_Count/Sum)*100
Female_Percentage = (Female_Count/Sum)*100
Non_Specific_Percentage = (Non_Specific/Sum)*100
```

```python
# Calculate percentage of attendees with a college degree
Attendance_Degree = (degree_holder / Population)*100
Attendance_Degree
```

```python
# Calculate average post-bootcamp salary
Salary = round(Attended.BootcampPostSalary.mean(), 2)
Salary
```

```python
# Create a new table consolidating above calculations
Summary_df = pd.DataFrame({'Total Surveyed':[Total],
                          'Total Bootcamp Attendees':[Attend],
                          '% attended Bootcamp':[Percentage],
                          'Avg. Age':[Average_Age],
                          'Has a Degree':[Attendance_Degree],
                          '% Male':[Male_Percentage],
                          '% Female':[Female_Percentage],
                          '% Non Gender Specific':[Non_Specific_Percentage],
                          'Average Post Bootcamp Salary':[Salary]})
Summary_df
```

```python
# Improve formatting before outputting spreadsheet
Summary_df["% Male"] = Summary_df["% Male"].map("{:,.2f}%".format)
Summary_df
```

```python
Summary_df["% Female"] = Summary_df["% Female"].map("{:,.2f}%".format)
Summary_df
```

```python
Summary_df["% attended Bootcamp"] = Summary_df["% attended Bootcamp"].map("{:,.2f}%".format)
Summary_df
```

```python
Summary_df["% Non Gender Specific"] = Summary_df["% Non Gender Specific"].map("{:,.2f}%".format)
Summary_df
```

```python
Summary_df['Average Post Bootcamp Salary'] = Summary_df['Average Post Bootcamp Salary'].map("${:,.2f}".format)
Summary_df
```

```python
# Export to Excel
# Use df.to_excel to export to excel. Don't include the indexes
Summary_df.to_excel(r'Output/Summary_df.xlsx', index=False)
```
