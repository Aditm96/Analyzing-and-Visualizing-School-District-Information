```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
# You may need to encode this file
File = pd.read_csv(file, low_memory = False)
File_df = pd.DataFrame(File)
File_df
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Consider using iloc
File_df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
```

```python
# Change reading 0 and 1 to No and Yes, respectively
File_df.replace(to_replace=[0.0, 1.0], value=["No", "Yes"], inplace = True)
File_df
```

```python
# Extract rows for only those who attended a bootcamp
Attended = File_df.loc[File_df["AttendedBootcamp"] == "Yes", :]
Attended
```

```python
# Create DataFrame of the various boot camps along with the number of participants (enrollees)
# Create a DataFrame from using value_counts
Bootcamp_name = pd.DataFrame(Attended.BootcampName.value_counts())
Bootcamp_name.reset_index(inplace=True)
Bootcamp_name.head()
Bootcamp_name = Bootcamp_name.rename(columns={"index":"BootcampName", "BootcampName":"Count"})
Bootcamp_name.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
Recommend = Attended.loc[Attended['BootcampRecommend'] == "Yes"]
Counts = Recommend.groupby("BootcampName")
Count = pd.DataFrame(Counts["BootcampRecommend"].count().astype(float))
Count.head()
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merge_df = pd.merge(Bootcamp_name, Count, on="BootcampName", how="outer")
merge_df_renamed = merge_df.rename(columns={"BootcampRecommend":"Recommenders"})
merge_df_renamed.head()
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
# Think of what should be the numerator and the demoninator
merge_df_renamed['% Recommend'] = (merge_df_renamed['Recommenders'] / merge_df_renamed['Count'])*100
merge_df_renamed["% Recommend"] = merge_df_renamed["% Recommend"].map("{:,.2f}%".format)
# Sort results in descending order
merge_df_renamed.sort_values(by=["Recommenders"], inplace=True, ascending=False)
merge_df_renamed
```

```python
# Export to excel and remove index
merge_df_renamed.to_excel(r'Output/Data_Clean_Up_Part_2.xlsx', index=False)
```
